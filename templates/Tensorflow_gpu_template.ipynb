```python
import tensorflow as tf
import pandas as pd
import numpy as np

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils, to_categorical
```

```python
tf.config.list_physical_devices()
```

```python
tf.__version__
```

```python
tf.debugging.set_log_device_placement(True)
tf.config.list_physical_devices()
```

```python
wine = load_wine()
```

```python
train_x, test_x, train_y, test_y = train_test_split(
    wine["data"],
    wine["target"],
    test_size=.33
)
```

```python
train_x = tf.convert_to_tensor(train_x, dtype=np.float32)
test_x = tf.convert_to_tensor(test_x, dtype=np.float32)
train_y = to_categorical(train_y)
test_y = to_categorical(test_y)
```

```python
model = Sequential()
model.add(Dense(9, input_dim=13, activation="relu"))
model.add(Dense(10, activation="relu"))
model.add(Dense(3, activation="softmax"))
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['acc'])
```

```python
model.fit(train_x, train_y, epochs=300, batch_size=10)
```

```python
scores = model.evaluate(test_x, test_y)
```

```python
print(f"Accuracy: {scores[1]*100}%")
```
